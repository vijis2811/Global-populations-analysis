```python
print("Total Countries:", g['Country'].nunique())
print("Shape of dataset:", g.shape)
```

```python
correlation = g.corr(numeric_only=True)
plt.figure(figsize=(10,6))
sns.heatmap(correlation, annot=True, cmap='coolwarm', fmt=".2f")
plt.title('Correlation Heatmap')
plt.show()
```

```python
sns.scatterplot(data=g, x='Population Aged 0 to 14 (%)', y='Population Aged 60 and Over (%)')
plt.title('Young vs Elderly Population')
plt.show()
```

```python
top10_pop = g.nlargest(10, 'Population(in millions)')
sns.barplot(x='Population(in millions)', y='Country', data=top10_pop)
plt.title('Top 10 Most Populated Countries')
plt.show()
```

```python
top10_dense = g.nlargest(10, 'Population density')
sns.barplot(x='Population density', y='Country', data=top10_dense)
plt.title('Top 10 Densely Populated Countries')
plt.show()
```

```python
top10_old = g.nlargest(10, 'Population Aged 60 and Over (%)')
sns.barplot(x='Population Aged 60 and Over (%)', y='Country', data=top10_old)
plt.title('Countries with Highest Elderly Population (%)')
plt.show()
```

```python
top10_pop = g.nlargest(10, 'Population(in millions)')
top10_pop_melt = top10_pop.melt(id_vars='Country',
                                value_vars=['Female Population(in millions)', 'Male Population(in millions)'],
                                var_name='Gender', value_name='Population')
plt.figure(figsize=(10,6))
sns.barplot(y='Country', x='Population', hue='Gender', data=top10_pop_melt)
plt.title('Gender Distribution in Top 10 Populated Countries', fontsize=14)
plt.show()
```

```python
g['Gender Gap'] = g['Male Population(in millions)'] - g['Female Population(in millions)']
top10_gap = g.nlargest(10, 'Gender Gap')
plt.figure(figsize=(10,6))
sns.barplot(y='Country', x='Gender Gap', data=top10_gap, palette='coolwarm')
plt.title('Countries with Largest Maleâ€“Female Population Gap', fontsize=14)
plt.xlabel('Difference (Males - Females, millions)')
plt.ylabel('')
plt.show()
```

```python
lowest_sex = g.nsmallest(10, 'Sex ratio (males per 100 females)')
plt.figure(figsize=(10,6))
sns.barplot(y='Country', x='Sex ratio (males per 100 females)', data=lowest_sex, palette='rocket')
plt.title('Countries with Lowest Male-to-Female Ratios', fontsize=14)
plt.xlabel('Males per 100 Females')
plt.ylabel('')
plt.show()
```

```python
fig, axes = plt.subplots(1, 3, figsize=(18,6))
sns.barplot(y='Country', x='Population(in millions)', data=g.nlargest(5, 'Population(in millions)'), ax=axes[0], palette='Blues')
axes[0].set_title('Top 5 by Total Population')
sns.barplot(y='Country', x='Population Aged 0 to 14 (%)', data=g.nlargest(5, 'Population Aged 0 to 14 (%)'), ax=axes[1], palette='Greens')
axes[1].set_title('Top 5 by Young Population')
sns.barplot(y='Country', x='Population Aged 60 and Over (%)', data=g.nlargest(5, 'Population Aged 60 and Over (%)'), ax=axes[2], palette='Oranges')
axes[2].set_title('Top 5 by Elderly Population')
plt.tight_layout()
plt.show()
```

```python
high_density = g.nlargest(10, 'Population density')
low_density = g.nsmallest(10, 'Population density')
fig, axes = plt.subplots(1, 2, figsize=(16,6))
sns.barplot(x='Population Aged 0 to 14 (%)', y='Country', data=high_density, ax=axes[0], palette='Blues')
axes[0].set_title('High-Density Nations (Youth %)')
sns.barplot(x='Population Aged 0 to 14 (%)', y='Country', data=low_density, ax=axes[1], palette='Greens')
axes[1].set_title('Low-Density Nations (Youth %)')
plt.tight_layout()
plt.show()
```

```python
top_female = g.nlargest(10, 'Female Population(in millions)')
plt.figure(figsize=(10,6))
sns.barplot(y='Country', x='Female Population(in millions)', data=top_female, palette='Purples_r')
plt.title('Top 10 Countries by Female Population')
plt.xlabel('Female Population (millions)')
plt.ylabel('')
plt.show()
```

```python
top_male = g.nlargest(10, 'Male Population(in millions)')
plt.figure(figsize=(10,6))
sns.barplot(y='Country', x='Male Population(in millions)', data=top_male, palette='Blues_r')
plt.title('Top 10 Countries by Male Population')
plt.xlabel('Male Population (millions)')
plt.ylabel('')
plt.show()
```

```python
g_small_pop_dense = g[g['Population(in millions)'] < 20].nlargest(10, 'Population density')
plt.figure(figsize=(10,6))
sns.barplot(y='Country', x='Population density', data=g_small_pop_dense, palette='flare')
plt.title('Small Countries with Extremely High Density (<20M Population)')
plt.xlabel('Population Density (people per kmÂ²)')
plt.ylabel('')
plt.show()
```

```python
g_low_density = g[g['Population density'] < 100].nlargest(10, 'Population(in millions)')
plt.figure(figsize=(10,6))
sns.barplot(y='Country', x='Population(in millions)', data=g_low_density, palette='crest')
plt.title('Top 10 Countries with High Population but Low Density')
plt.xlabel('Population (millions)')
plt.ylabel('')
plt.show()
```

```python
top_youth = g.nlargest(10, 'Population Aged 0 to 14 (%)')
plt.figure(figsize=(10,6))
sns.barplot(y='Country', x='Population Aged 0 to 14 (%)', data=top_youth, palette='Greens_r')
plt.title('Countries with Highest Youth Population (%)')
plt.xlabel('% Population aged 0â€“14')
plt.ylabel('')
plt.show()
```

```python
top_old = g.nlargest(10, 'Population Aged 60 and Over (%)')
plt.figure(figsize=(10,6))
sns.barplot(y='Country', x='Population Aged 60 and Over (%)', data=top_old, palette='magma')
plt.title('Countries with Highest Elderly Population (%)')
plt.xlabel('% Population aged 60+')
plt.ylabel('')
plt.show()
```

```python
g_old_small = g[g['Population(in millions)'] < 20].nlargest(10, 'Population Aged 60 and Over (%)')
plt.figure(figsize=(10,6))
sns.barplot(y='Country', x='Population Aged 60 and Over (%)', data=g_old_small, palette='plasma')
plt.title('Aging Small Nations (High Elderly %, Pop < 20M)')
plt.xlabel('Elderly Population (%)')
plt.ylabel('')
plt.show()
```

```python
g['Global_Share_%'] = (g['Population(in millions)'] / g['Population(in millions)'].sum()) * 100
top_share = g.nlargest(10, 'Global_Share_%')
plt.figure(figsize=(8,6))
sns.barplot(y='Country', x='Global_Share_%', data=top_share, palette='cubehelix')
plt.title('Top 10 Countries by Share of Global Population')
plt.xlabel('Global Population Share (%)')
plt.ylabel('')
plt.show()
```

```python
g['Gender_Diff_%'] = abs(g['Male Population(in millions)'] - g['Female Population(in millions)']) / g['Population(in millions)'] * 100
top_gender_diff = g.nlargest(15, 'Gender_Diff_%')
plt.figure(figsize=(10,6))
sns.barplot(y='Country', x='Gender_Diff_%', data=top_gender_diff, palette='RdYlBu')
plt.title('Top 15 Countries by Gender Imbalance (%)')
plt.xlabel('Gender Imbalance (% of Total Population)')
plt.ylabel('')
plt.show()
```
